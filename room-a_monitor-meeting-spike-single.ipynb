## Clean-up

### Load the full CSV

```python
def import_csv(file):
    df = pd.read_csv(file)
    return df

file = 'monitor-data/room-a_monitor_meeting_spike_data.csv'

full_data = import_csv(file)

print(full_data);
```

### Remove irrelevant rows

Remove rows the other sensory data irrelevant for this analysis.

```python
def filter_columns(data):
    data.columns = data.columns.str.strip()
    filtered_df = data[['a Date', 'Time', 'CO2']]
    return filtered_df

filtered_data = filter_columns(full_data)
print(filtered_data)
```

### Plot the CO2 distribution

```python
print(filtered_data)

# Plot CO2 levels over time
plt.figure(figsize=(10, 6))
plt.plot(filtered_data.index, filtered_data['CO2'], linestyle='-', color='#7168bc', label='CO2 concentrations')

# Plot bars at every data point
plt.bar(filtered_data.index, filtered_data['CO2'], color='#deddee', width=1, edgecolor='#ada9d6',)

# Draw horizontal line at 600 CO2 level mark
plt.axhline(y=600, color='#756db8', linestyle='--', label='CO2 level treshold')

title = plt.title('Room A - Single Meeting / IAQ Monitor (03-04-24)'.upper(), fontweight="bold" )
title.set_position([.5, 0.6])  # Adjust the title position
title.set_color('white')  # Set title text color to white
title.set_bbox(dict(facecolor='black', edgecolor='black', boxstyle='round,pad=0.3'))  # Set title background color to black
plt.xlabel('Time of day (${min}$)') 
plt.ylabel('CO2 Level (${ppm}$)')

# Set y-axis ticks at increments of 50
plt.yticks(range(0, int(filtered_data['CO2'].max())+1, 25))

# Set y-axis limit
plt.ylim(450, 750)

# Set x-axis ticks at intervals of 5 minutes
tick_positions = range(0, len(filtered_data), 5)
plt.xticks(tick_positions, filtered_data['Time'][tick_positions], rotation=45)

# Add markers at the 5-minute intervals
plt.plot(filtered_data.index[tick_positions], filtered_data['CO2'].iloc[tick_positions], markersize=4, linestyle='', color='#407ab1')

# Reverse the x-axis
plt.gca().invert_xaxis()

# Add vertical lines for start and end of meeting
start_index = filtered_data[filtered_data['Time'] == '10:15'].index[0]
end_index = filtered_data[filtered_data['Time'] == '11:30'].index[0]
plt.axvline(x=start_index, color='#84b690', linestyle='--', label='Meeting start: 10:15')
plt.axvline(x=end_index, color='#d85c36', linestyle='--', label='Meeting end: 11:30')

plt.grid(True, linestyle='--', color='lightgrey')
plt.tight_layout()
plt.legend()

# Save the chart to disk
plt.savefig('monitor-charts/room-a-monitor-meeting-spike.png', dpi=300, bbox_inches='tight')

plt.show()
```
